**SED Lineales - caso Reale y diferentes**

**Algoritmo Solución**

*   DET $(A-I\lambda)$
*   Calculo de polinomio Catacteritico 
*   Valores Propios
*   Evaluar $(A-I\lambda)\vec{V}=\vec{0}$ y encontrar $\vec{V}$
*   Plantear Respuesta 

$\vec{X}=c_1\vec{V_1}e^{\lambda_1t}+c_2\vec{V_2}e^{\lambda_2t}$

```python
from sympy import *

A=Matrix([[-3,-2],
          [-1,-2]])
A
```

```python
val_p=A.eigenvals()
list(val_p)
```

```python
# solucionando para el primer valor de lamba
l=list(val_p)[0]

a,b=symbols("a,b")

V=Matrix([[a],
          [b]])

expr=Eq((A-eye(2)*l)*V,zeros(2,1))
expr

ics={b:1}  #diccionario
 
solve(expr,ics=ics)

b=1
v_1=Matrix([[-b],
            [b]])
print(f"lambda={l}")
print(f"v1=")
v_1
```

```python
# solucionando para el segundo valor de lamba
l=list(val_p)[1]   #valor de lambda

a,b=symbols("a,b")

V=Matrix([[a],
          [b]])

expr=Eq((A-eye(2)*l)*V,zeros(2,1))
expr

solve(expr,ics={b:1})
b=1
v_2=Matrix([[2*b],
            [b]])

print(f"lambda={l}")
print(f"v_2=")
v_2
```
