```python
## if 
salario=int(input("salario:"))
salario_minimo=1000000

if salario>salario_minimo:
  gasto=int(input("Gastos:"))
  print("Usted gana mas del salario minimo y sus ahorros son de {}".format(salario-gasto))
elif salario==salario_minimo:
  gasto=int(input("Gastos:"))
  print(f"Usted gana excatamente el salario minimo y sus ahorros son de {salario-gasto}")
else:
  gasto=int(input("Gastos:"))
  print(f"Ustede gana menos del salario minimo y sus ahorros son de {salario-gasto}")

```

```python
numero1=40
numero2=21

if numero1%2 == 0 and numero2%2 == 0:
  print("Los dos numeros son pares")
elif numero1%2 == 0 or numero2%2 == 0:
  if numero1%2 == 0:
    print("El numero 1 es par")
  if numero2%2 == 0:
    print("El numero 2 es par")
else:
  print("Ninguno de los dos numeros es par ")
```

#### **List comprehension**

Vamos a crear listas por medio de bucles 

```python

lista_pares=[i for i in range(0,25) if i%2==0]   #[elemento_nuevo for elemento in iterable_elementos if sentencia]
lista_pares
```

```python
lista_pares_cuadrado=[i**2 for i in range(0,25) if i%2==0]
lista_pares_cuadrado
```

```python
salarios=[15,20,26,45,80,65]
salario_minimo=35

menos_minimo=[i for i in salarios if i<salario_minimo]
menos_minimo
```

```python
# descripcion de la funcion 
def impares(a):
  '''Esta Funcion cuenta los numeros 
  impares de una lista'''
  contador=0
  for numero in a:
    if numero%2!=0:
      contador+=1
    else:
      None
  return contador  
```

```python
lista_test=[4,7,8,9,3,8]
impares(lista_test)
```
